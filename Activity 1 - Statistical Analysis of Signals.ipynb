# Activity 1 - Statistical Analysis of Signals

Filter files from the detaset using glob with regex.

```python
import glob
steps = glob.glob('./dataset/G1_Hd1_FTDS_SP1_050418_162003_1_0_*[0-9]_B1.csv') # Get steps with wildcard 0-9 for B1 using regex.
print(steps) # Preview the list of gathered files.
print("Number of files: {}".format(len(steps))) # Count the number of files.
```

Import all necessary the libraries for analysis.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Create a for loop to do analysis for each step.

```python
df = pd.read_csv(steps[0], skiprows=5) # Skip the first five rows in the csv file.
df = df[['AE [V]']] # Restructure the dataframe to get only the 'AE [V]' column.
df.head(15) # Preview the first 15 entries in the file.
df.info()

entries_count = df.count()['AE [V]'] # Get number of entries from the 'AE [V]' column.
mean = df.mean()['AE [V]'] # Get mean of entries from the 'AE [V]' column.
std_dev = df.std()['AE [V]'] # Get standard deviation of entries from the 'AE [V]' column.
std_err = df.sem()['AE [V]'] # Get standard error of entries from the 'AE [V]' column.

print("Number of entries: {}".format(entries_count)) # Display the number of entries.
print("Mean: {}".format(mean)) # Display the mean.
print("Standard Deviation: {}".format(std_dev)) # Display the standard deviation.
print("Standard Error: {}".format(std_err)) # Display the standard error.
```

```python
# Plotting of histogram.
df['AE [V]'].plot(kind='hist', bins=2500, title="Step #")
```

```python
# Plotting of means.
df['AE [V]'].plot()
```

```python
for step in steps:
    df = pd.read_csv(step, skiprows=5) # Skip the first five rows in the csv file.
    df = df[['AE [V]']] # Restructure the dataframe to get only the 'AE [V]' column.
    df.head(15) # Preview the first 15 entries in the file.
    
    step_num = steps.index(step) # Get the current step from index.
    entries_count = df.count()['AE [V]'] # Get number of entries from the 'AE [V]' column.
    mean = df.mean()['AE [V]'] # Get mean of entries from the 'AE [V]' column.
    std_dev = df.std()['AE [V]'] # Get standard deviation of entries from the 'AE [V]' column.
    std_err = df.sem()['AE [V]'] # Get standard error of entries from the 'AE [V]' column.
    
    print("Step {}".format(step_num)) # Display the current step.
    print("Number of entries: {}".format(entries_count)) # Display the number of entries.
    print("Mean: {}".format(mean)) # Display the mean.
    print("Standard Deviation: {}".format(std_dev)) # Display the standard deviation.
    print("Standard Error: {}".format(std_err)) # Display the standard error.
    print("\n".format(step_num)) # Separator for each display iteration.
    
#     # Time-series plot.
#     time_series = df['AE [V]'].plot(title='Step #{}'.format(step_num))
    
#     # Plotting of histogram.
#     histogram = df['AE [V]'].plot(kind='hist', bins=2500, title="Step #")
```
